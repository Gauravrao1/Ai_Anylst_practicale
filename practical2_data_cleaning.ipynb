##### Practical 2: Data Cleaning and Organization of Supermarket Data
Dataset: Superstore Sales Dataset

###### 1. Import Libraries and Load Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_excel("superstore_sales_cleaned.xlsx")
print("Shape:", df.shape)
df.head()
```

###### 2. Check Data Types and Missing Values

```python
print("Data Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

###### 3. Handle Null Values

```python
# Fill numeric nulls with mean, categorical with mode
for col in df.select_dtypes(include=[np.number]).columns:
    df[col].fillna(df[col].mean(), inplace=True)

for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("Null values after handling:", df.isnull().sum().sum())
```

###### 4. Remove Duplicates

```python
print("Duplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates after:", df.duplicated().sum())
```

###### 5. Fix Column Types

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Postal_Code'] = df['Postal_Code'].astype(str)

print("Updated Data Types:\n", df.dtypes)
```

###### 6. Format Dates

```python
df['Order_Date_Formatted'] = df['Order_Date'].dt.strftime('%d-%m-%Y')
df['Ship_Date_Formatted'] = df['Ship_Date'].dt.strftime('%d-%m-%Y')

print("Formatted dates:")
df[['Order_Date', 'Order_Date_Formatted', 'Ship_Date', 'Ship_Date_Formatted']].head()
```

###### 7. Verify and Save Cleaned Data

```python
print("Final Shape:", df.shape)
print("Null Values:", df.isnull().sum().sum())
df.to_excel("superstore_sales_final_cleaned.xlsx", index=False)
print("Saved as superstore_sales_final_cleaned.xlsx")
df.head()
```
