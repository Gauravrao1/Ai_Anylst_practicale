##### Practical 3: Analyze Monthly Sales Trends Using Visualizations
Dataset: Superstore Sales Dataset

###### 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
print("Libraries imported successfully")
```

###### 2. Load and Prepare Data

```python
df = pd.read_excel("superstore_sales_cleaned.xlsx")
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Month'] = df['Order_Date'].dt.month_name()
df['Month_Num'] = df['Order_Date'].dt.month

# Monthly sales summary
monthly_sales = df.groupby(['Month_Num', 'Month'])['Sales'].sum().reset_index()
monthly_sales = monthly_sales.sort_values('Month_Num')
print("Monthly Sales Summary:")
monthly_sales
```

###### 3. Bar Plot - Monthly Sales

```python
plt.figure(figsize=(10, 5))
plt.bar(monthly_sales['Month'], monthly_sales['Sales'], color='skyblue', edgecolor='black')
plt.title('Monthly Sales - Bar Plot')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###### 4. Pie Chart - Monthly Sales Distribution

```python
plt.figure(figsize=(8, 8))
plt.pie(monthly_sales['Sales'], labels=monthly_sales['Month'], autopct='%1.1f%%', startangle=140)
plt.title('Monthly Sales Distribution - Pie Chart')
plt.tight_layout()
plt.show()
```

###### 5. Line Graph - Monthly Sales Trend

```python
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales['Month'], monthly_sales['Sales'], marker='o', color='green', linewidth=2)
plt.title('Monthly Sales Trend - Line Graph')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

###### 6. Category-wise Sales - Bar Plot

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
category_sales.plot(kind='bar', color=['coral', 'steelblue', 'gold'], edgecolor='black')
plt.title('Category-wise Sales - Bar Plot')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

###### 7. Region-wise Sales - Pie Chart

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(7, 7))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90,
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Region-wise Sales Distribution - Pie Chart')
plt.tight_layout()
plt.show()
```
