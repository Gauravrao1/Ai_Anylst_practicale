##### Practical 4: Visualization and Exploratory Data Analysis (EDA)
Dataset: House Prices Dataset

###### 1. Import Libraries and Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dirty_house_prices.csv")
df.dropna(inplace=True)
print("Shape:", df.shape)
df.head()
```

###### 2. Bar Plot - Average Price by Neighborhood

```python
avg_price = df.groupby('Neighborhood')['SalePrice'].mean()
avg_price.plot(kind='bar', color='skyblue', edgecolor='black', figsize=(8, 4))
plt.title('Average Sale Price by Neighborhood')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

###### 3. Pie Chart - House Style Distribution

```python
df['HouseStyle'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title('House Style Distribution')
plt.ylabel('')
plt.show()
```

###### 4. Histogram - Sale Price Distribution

```python
df['SalePrice'].plot(kind='hist', bins=20, color='coral', edgecolor='black', figsize=(8, 4))
plt.title('Sale Price Distribution')
plt.xlabel('Sale Price')
plt.tight_layout()
plt.show()
```

###### 5. Line Plot - Year Built vs Sale Price

```python
year_avg = df.groupby('YearBuilt')['SalePrice'].mean().sort_index()
year_avg.plot(kind='line', marker='o', figsize=(10, 4), color='green')
plt.title('Average Sale Price by Year Built')
plt.xlabel('Year Built')
plt.ylabel('Average Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

###### 6. Scatter Plot - Lot Area vs Sale Price

```python
plt.figure(figsize=(8, 4))
plt.scatter(df['LotArea'], df['SalePrice'], alpha=0.5, color='purple')
plt.title('Lot Area vs Sale Price')
plt.xlabel('Lot Area')
plt.ylabel('Sale Price')
plt.tight_layout()
plt.show()
```

###### 7. Box Plot - Sale Price by House Style

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='HouseStyle', y='SalePrice', data=df)
plt.title('Sale Price by House Style')
plt.tight_layout()
plt.show()
```

###### 8. Heatmap - Correlation Matrix

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

###### 9. Pair Plot - EDA Overview

```python
sns.pairplot(df[['LotArea', 'NumRooms', 'GarageArea', 'SalePrice']])
plt.suptitle('Pair Plot - Key Features', y=1.02)
plt.show()
```

###### 10. Seaborn - Violin Plot (Price by Neighborhood)

```python
plt.figure(figsize=(8, 4))
sns.violinplot(x='Neighborhood', y='SalePrice', data=df)
plt.title('Violin Plot - Sale Price by Neighborhood')
plt.tight_layout()
plt.show()
```

###### 11. Seaborn - Count Plot (House Style Count)

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='HouseStyle', data=df, palette='Set2')
plt.title('Count Plot - House Style')
plt.tight_layout()
plt.show()
```

###### 12. Seaborn - KDE Plot (Sale Price Density)

```python
plt.figure(figsize=(8, 4))
sns.kdeplot(df['SalePrice'], fill=True, color='orange')
plt.title('KDE Plot - Sale Price Density')
plt.xlabel('Sale Price')
plt.tight_layout()
plt.show()
```
