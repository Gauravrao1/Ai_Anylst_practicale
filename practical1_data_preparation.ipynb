##### Practical 1: Data Preparation and Conversion Using Python (Pandas)
Dataset: Student Performance Dataset (2000 records)

###### 1. Import Libraries

```python
import pandas as pd
import numpy as np
print("Libraries imported successfully")
```

###### 2. Load the Dataset

```python
df = pd.read_csv("student_performance_2000 (1).csv")
print("Shape of dataset:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Dataset info and summary statistics
print("Dataset Info:")
print(df.info())
print("\nStatistical Summary:")
df.describe()
```

###### 3. Check Missing Values

```python
print("Missing values in each column:\n")
print(df.isnull().sum())
print("\nTotal missing values:", df.isnull().sum().sum())
```

###### 4. Handle Missing Values

```python
# Fill numeric columns with mean
df['Attendance_Percentage'].fillna(df['Attendance_Percentage'].mean(), inplace=True)
df['Internal_Marks'].fillna(df['Internal_Marks'].mean(), inplace=True)
df['External_Marks'].fillna(df['External_Marks'].mean(), inplace=True)

# Fill categorical column with mode
df['Department'].fillna(df['Department'].mode()[0], inplace=True)

print("Missing values after handling:\n")
print(df.isnull().sum())
```

###### 5. Check and Remove Duplicates

```python
print("Duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", df.shape)
```

###### 6. Data Type Conversion

```python
print("Data types BEFORE conversion:\n")
print(df.dtypes)

# Fill remaining missing values in Age and City
df['Age'].fillna(df['Age'].median(), inplace=True)
df['City'].fillna(df['City'].mode()[0], inplace=True)

# Convert columns to appropriate types
df['Age'] = df['Age'].astype(int)
df['Semester'] = df['Semester'].astype(int)
df['Gender'] = df['Gender'].astype('category')
df['Department'] = df['Department'].astype('category')

print("\nData types AFTER conversion:\n")
print(df.dtypes)
```

###### 7. Rename Columns

```python
df.rename(columns={
    'Attendance_Percentage': 'Attendance',
    'Internal_Marks': 'Internal',
    'External_Marks': 'External'
}, inplace=True)

print("Columns after renaming:\n")
print(df.columns.tolist())
```

###### 8. Feature Engineering

```python
# Create Total Marks column
df['Total_Marks'] = df['Internal'] + df['External']

# Create Result column (Pass if Total >= 50, else Fail)
df['Result'] = np.where(df['Total_Marks'] >= 50, 'Pass', 'Fail')

print("New columns added: Total_Marks, Result\n")
print(df[['Name', 'Internal', 'External', 'Total_Marks', 'Result']].head(10))
```

###### 9. Data Filtering

```python
# Students who failed
failed = df[df['Result'] == 'Fail']
print(f"Total failed students: {len(failed)}\n")
print(failed[['Name', 'Department', 'Total_Marks', 'Result']].head())

# Filter by department
print(f"\n\nStudents in CSE department: {len(df[df['Department'] == 'CSE'])}")
print(f"Students in IT department: {len(df[df['Department'] == 'IT'])}")
```

###### 10. Save Cleaned Data

```python
df.to_csv("cleaned_student_data.csv", index=False)
print("Cleaned data saved to 'cleaned_student_data.csv'")
print("\nFinal dataset shape:", df.shape)
print("\nFinal columns:", df.columns.tolist())
df.head()
```
